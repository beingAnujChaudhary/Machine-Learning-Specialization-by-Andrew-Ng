## Question 1

### ❓ Which of these best describes **unsupervised learning**?

* A form of machine learning that finds patterns using unlabeled data (x).
* A form of machine learning that finds patterns using labeled data (x, y)
* A form of machine learning that finds patterns without using a cost function.
* A form of machine learning that finds patterns in data using only labels (y) but without any inputs (x).

---

### ✅ Correct Answer:

* **A form of machine learning that finds patterns using unlabeled data (x).**

---

### 🧠 Explanation of Correct Answer:

**Unsupervised learning** is a type of machine learning where:

* We only have **input data (x)**
* There are **no labels (y)**
* The goal is to discover hidden patterns or structures in the data

👉 The model learns *on its own* without being told the correct answers.

---

### ❌ Why the Other Options Are Incorrect:

* **Labeled data (x, y)** → This describes **supervised learning**, not unsupervised learning
* **Without using a cost function** → Incorrect, because most algorithms (like K-means) *do use a cost function*
* **Only labels (y)** → Impossible scenario (no input features to learn from)

---

### 🌍 Real-World Examples:

* **Clustering customers** based on shopping behavior (no labels given)
* **Grouping news articles** by topic automatically

## Question 2

### ❓ Which statements are true about **K-means**? (Select all that apply)

* The number of cluster centroids μₖ is equal to the number of examples.
* The number of cluster assignment variables c(i) is equal to the number of training examples.
* If K = 3, then each c(i) should be 1, 2, or 3.
* If each example x has 5 features, then each centroid μₖ also has 5 features.

---

### ✅ Correct Answers:

* **The number of cluster assignment variables c(i) is equal to the number of training examples.**
* **If K = 3, then each c(i) should be 1, 2, or 3.**
* **If each example x has 5 features, then each centroid μₖ also has 5 features.**

---

### 🧠 Explanation of Correct Answers:

1. **c(i) = number of training examples**

   * Each data point must belong to one cluster
   * So we need one assignment per example

2. **Cluster labels range from 1 to K**

   * If K = 3 → clusters are labeled {1, 2, 3}

3. **Centroid dimensions = input dimensions**

   * If x ∈ ℝ⁵ → centroid must also be in ℝ⁵

---

### ❌ Why the Other Option Is Incorrect:

* **Centroids equal to number of examples** ❌

  * Wrong: number of centroids = K (chosen by us), not m (dataset size)

---

### 🌍 Real-World Example:

* Grouping **customers into 3 segments** (K=3)
* Each customer has 5 features → centroid also has 5 features

## Question 3

### ❓ After running K-means 100 times, how should you pick the best solution?

* Pick the one with the lowest cost J
* Pick randomly
* Pick the last one
* Average all solutions

---

### ✅ Correct Answer:

* **Pick the one with the lowest cost J**

---

### 🧠 Explanation:

K-means can get stuck in **local minima**, so:

* Different initializations → different results
* We choose the solution with **minimum cost function**

---

### ❌ Incorrect Options:

## Question 4

### ❓ The **elbow method** is used to:

* Choose the number of clusters K
* Choose max examples per cluster
* Choose best initialization
* Choose dataset size

---

### ✅ Correct Answer:

* **Choose the number of clusters K**

---

### 🧠 Explanation:

* Plot **cost J vs number of clusters K**
* Look for the “elbow point” where improvement slows down

👉 That point gives the optimal K

---

### 💻 Visualization Example:

```python
import matplotlib.pyplot as plt

K = [1, 2, 3, 4, 5]
cost = [100, 60, 40, 35, 33]

plt.plot(K, cost, marker='o')
plt.xlabel("Number of clusters (K)")
plt.ylabel("Cost (J)")
plt.title("Elbow Method")
plt.show()
```

---

### ❌ Incorrect Options:

* Not about dataset size
* Not about initialization
* Not about cluster capacity

---

### 🌍 Real-World Example:

* Choosing how many **customer segments** to create

## Question 5

### ❓ What happens to the cost function J during K-means iterations?

* The cost decreases or stays the same
* The cost fluctuates randomly
* There is no cost function
* K-means maximizes cost

---

### ✅ Correct Answer:

* **The cost will either decrease or stay the same after each iteration**

---

### 🧠 Explanation:

K-means has two steps:

1. **Assign clusters** → reduces distance
2. **Update centroids** → improves cluster centers

👉 Each step ensures cost **never increases**

---

### ❌ Incorrect Options:

* **Fluctuates randomly** ❌ → K-means is monotonic
* **No cost function** ❌ → clearly defined
* **Maximizes cost** ❌ → it *minimizes* cost

---

### 🌍 Real-World Example:

* Each iteration → customers grouped better → tighter clusters
