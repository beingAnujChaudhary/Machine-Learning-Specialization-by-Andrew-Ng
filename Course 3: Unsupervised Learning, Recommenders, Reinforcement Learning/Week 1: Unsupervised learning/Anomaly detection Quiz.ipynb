## Question 1

**Question Presentation:**
You are building a system to detect if computers in a data center are malfunctioning. You have 10,000 data points of computers functioning well, and **no data** from computers malfunctioning. What type of algorithm should you use?

*   Anomaly detection
*   Supervised learning

### Correct Answer
**Anomaly detection**

### Explanation
**Anomaly Detection** is the correct choice here because you have a highly imbalanced dataset where you only have examples of the "normal" class (functioning well).
*   **Concept:** Anomaly detection algorithms work by learning the characteristics of "normal" data. Once the model understands what "normal" looks like, it can flag any new data point that deviates significantly from this norm as an anomaly (malfunction).
*   **Why not Supervised Learning?** Supervised learning (specifically classification) requires labeled examples of **both** classes (malfunctioning and functioning) to learn the boundary between them. Since you have zero examples of malfunctioning computers, a supervised algorithm has nothing to learn about the "malfunction" class.

### Real-World Example
*   **Fraud Detection:** A bank has millions of records of legitimate transactions but very few records of fraud. They use anomaly detection to learn normal spending patterns and flag unusual transactions.

## Question 2

**Question Presentation:**
You are building a system to detect if computers in a data center are malfunctioning. You have 10,000 data points of computers functioning well, and **10,000 data points** of computers malfunctioning. What type of algorithm should you use?

*   Anomaly detection
*   Supervised learning

### Correct Answer
**Supervised learning**

### Explanation
**Supervised Learning** is the correct choice because you have a balanced dataset with ample examples of both positive (malfunctioning) and negative (functioning) cases.
*   **Concept:** With labeled data for both classes, a supervised learning algorithm (like Logistic Regression or a Neural Network) can learn a decision boundary that explicitly separates functioning computers from malfunctioning ones. This is generally more powerful than anomaly detection when sufficient data is available.
*   **Why not Anomaly Detection?** While you *could* use anomaly detection, supervised learning is preferred when you have enough data for both classes because it utilizes the specific information about what malfunctions look like, rather than just treating them as "not normal."

### Real-World Example
*   **Spam Email Filter:** You have a dataset containing thousands of emails labeled as "Spam" and thousands labeled as "Not Spam." A supervised learning algorithm learns the specific words and patterns associated with spam.

## Question 3

**Question Presentation:**
Say you have 5,000 examples of normal airplane engines, and 15 examples of anomalous engines. How would you use the 15 examples of anomalous engines to evaluate your anomaly detection algorithm?

*   Put the data of anomalous engines (together with some normal engines) in the cross-validation and/or test sets to measure if the learned model can correctly detect anomalous engines.
*   Because you have data of both normal and anomalous engines, don't use anomaly detection. Use supervised learning instead.
*   Use it during training by fitting one Gaussian model to the normal engines, and a different Gaussian model to the anomalous engines.
*   You cannot evaluate an anomaly detection algorithm because it is an unsupervised learning algorithm.

### Correct Answer
**Put the data of anomalous engines (together with some normal engines) in the cross-validation and/or test sets to measure if the learned model can correctly detect anomalous engines.**

### Explanation
Even though anomaly detection is often considered "unsupervised" during the *training* phase (because it primarily learns from normal data), you still need labeled data to **evaluate** its performance.
*   **Evaluation Strategy:** You train the model only on the 5,000 normal engines. Then, you create a Cross-Validation (CV) or Test set that contains a mix of normal engines and the 15 anomalous engines.
*   **Goal:** You check if the model correctly flags the 15 anomalous engines as anomalies while keeping the false alarm rate (flagging normal engines as anomalies) low.

### Why other options are incorrect
*   **Supervised Learning:** With only 15 anomalous examples, the dataset is too imbalanced for effective supervised learning. The model would likely be biased toward the majority class (normal).
*   **Fitting Gaussian to Anomalies:** Anomaly detection typically fits a model (like a Gaussian) only to the *normal* data ($p(x)$). Fitting a model to the tiny set of 15 anomalies would result in a poor statistical model.
*   **Cannot Evaluate:** This is false. All machine learning models must be evaluated on held-out data to ensure they work.

## Question 4

**Question Presentation:**
Anomaly detection flags a new input $x$ as an anomaly if $p(x) < \epsilon$. If we reduce the value of $\epsilon$, what happens?

*   The algorithm is more likely to classify new examples as an anomaly. *(Note: This was selected in the image but is incorrect based on the 0/20 score)*
*   The algorithm is less likely to classify new examples as an anomaly.
*   The algorithm is more likely to classify some examples as an anomaly, and less likely to classify some examples as an anomaly. It depends on the example $x$.
*   The algorithm will automatically choose parameters $\mu$ and $\sigma$ to decrease $p(x)$ and compensate.

### Correct Answer
**The algorithm is less likely to classify new examples as an anomaly.**

### Explanation
The condition for flagging an anomaly is: **Flag if Probability $p(x)$ is LESS THAN Threshold $\epsilon$**.
*   **Logic:**
    *   Imagine your threshold $\epsilon$ is currently **0.5**. Any data point with a probability lower than 0.5 (e.g., 0.4, 0.3, 0.1) gets flagged.
    *   Now, you **reduce** $\epsilon$ to **0.1**.
    *   The data point with probability 0.4 is no longer less than 0.1. It is **not** flagged.
    *   Only data points with extremely low probability (e.g., 0.05) will now be flagged.
*   **Conclusion:** By lowering the threshold, you are making the criteria for being an anomaly **stricter**. Therefore, fewer examples will be classified as anomalies.

*(Note: The screenshot shows the first option selected, but the score is 0/20, indicating that selection was incorrect.)*

### Real-World Analogy
Think of $\epsilon$ as a "suspicion level."
*   High $\epsilon$: You are very suspicious. You flag anyone who acts slightly weird. (High False Positives).
*   Low $\epsilon$: You are very trusting. You only flag someone if they act extremely weird. (Low False Positives, but might miss some anomalies).

## Question 5

**Question Presentation:**
You are monitoring the temperature and vibration intensity on newly manufactured aircraft engines. You have measured 100 engines and fit the Gaussian model described in the video lectures to the data. The 100 examples and the resulting distributions are shown in the figure below.
The measurements on the latest engine you are testing have a temperature of 17.5 and a vibration intensity of 48. These are shown in magenta on the figure below. What is the probability of an engine having these two measurements?

image.png

*   $0.0738 \times 0.02288 = 0.00169$
*   $17.5 \times 48 = 840$
*   $0.0738 + 0.02288 = 0.0966$
*   $17.5 + 48 = 65.5$

### Correct Answer
**$0.0738 \times 0.02288 = 0.00169$**

### Explanation
In the standard anomaly detection model described (assuming features are independent), the probability of a new example $x$ is the product of the probabilities of its individual features.
*   **Formula:**
    $$p(x) = p(x_1; \mu_1, \sigma_1^2) \times p(x_2; \mu_2, \sigma_2^2) \times \dots \times p(x_n; \mu_n, \sigma_n^2)$$
    Or simply:
    $$p(x) = p(x_1) \times p(x_2)$$
*   **From the Graphs:**
    *   For Temperature ($x_1 = 17.5$), the graph shows the probability density $p(x_1) \approx 0.0738$.
    *   For Vibration ($x_2 = 48$), the graph shows the probability density $p(x_2) \approx 0.02288$.
*   **Calculation:**
    $$p(x) = 0.0738 \times 0.02288 \approx 0.00169$$

### Why other options are incorrect
*   **Multiplication of values (17.5 * 48):** This multiplies the raw feature values, which has no statistical meaning in this context.
*   **Addition of probabilities:** Probabilities of independent events occurring together are multiplied, not added. Adding them would imply an "OR" condition, not an "AND" condition (having BOTH this temperature AND this vibration).
*   **Addition of values:** Simply adding the sensor readings is not how probability is calculated.
