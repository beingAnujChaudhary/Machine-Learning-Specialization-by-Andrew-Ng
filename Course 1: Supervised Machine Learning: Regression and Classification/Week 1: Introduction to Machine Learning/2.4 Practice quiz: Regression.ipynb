## Question 1

**For linear regression, the model is**  
$f_{w,b}(x) = wx + b$.

**Which of the following are the inputs, or features, that are fed into the model and with which the model is expected to make a prediction?**

### Answer Options

- x
- w and b
- m
- (x, y)


### ✅ Correct Answer

**x**


### Explanation of the Correct Answer

In **Linear Regression**, the model is defined as:

$f_{w,b}(x) = wx + b$

Where:

- **x** = the **input feature** (the data given to the model)
- **w** = the **weight parameter** learned during training
- **b** = the **bias parameter** learned during training
- **f(x)** = the **predicted output**

The model takes **x as input** and produces a prediction.

So:

1. The **feature** provided to the model is **x**.
2. The model uses the learned parameters **w** and **b** to compute the prediction.
3. The result is the predicted value of **y**.

Therefore, **x is the correct answer because it is the feature fed into the model.**


### ❌ Why the Other Options Are Incorrect

**Option: w and b**

- **w (weight)** and **b (bias)** are **model parameters**, not inputs.
- They are **learned during training** using an optimization algorithm such as **gradient descent**.
- They help determine the slope and intercept of the regression line.

---

**Option: m**

- **m** usually represents the **number of training examples** in a dataset.
- It is not a feature or input to the model.
- Instead, it appears in formulas such as the **cost function**.

---

**Option: (x, y)**

- **x** is the input feature.
- **y** is the **target value (label)**.
- During training we use **(x, y)** pairs to teach the model.
- However, **only x is fed into the model to make predictions**.

## Question 2

**For linear regression, if you find parameters** $w$ **and** $b$ **so that** $J(w,b)$ **is very close to zero, what can you conclude?**

### Answer Options

- The selected values of the parameters w and b cause the algorithm to fit the training set really well.
- The selected values of the parameters w and b cause the algorithm to fit the training set really poorly.
- This is never possible --- there must be a bug in the code.


### ✅ Correct Answer

**The selected values of the parameters w and b cause the algorithm to fit the training set really well.**


### Explanation of the Correct Answer

In Linear Regression, the **cost function** is usually defined as:

$J(w,b) = \frac{1}{2m} \sum_{i=1}^{m} (f_{w,b}(x^{(i)}) - y^{(i)})^2$

Where:

- **m** = number of training examples
- **f(x)** = predicted value
- **y** = actual value

The cost function measures **prediction error**.

If **J(w,b) is close to zero**, it means:

1. The predictions are **very close to the true values**.
2. The model has **very small error**.
3. The regression line fits the training data **very well**.

Therefore, the chosen parameters **w and b represent a good model fit**.


### ❌ Why the Other Options Are Incorrect

**Option: The selected values cause the algorithm to fit the training set really poorly.**

- This would happen if the **cost function is very large**, not close to zero.
- A large cost means predictions are far from actual values.

---

**Option: This is never possible — there must be a bug in the code.**

- It is absolutely possible for the cost to approach **zero**.
- This happens when the regression line **perfectly fits the data**.
- For example, if all points lie exactly on a straight line.
