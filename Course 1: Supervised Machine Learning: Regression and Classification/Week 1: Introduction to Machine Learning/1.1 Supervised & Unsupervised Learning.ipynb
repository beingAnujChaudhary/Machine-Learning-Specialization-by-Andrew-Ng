# Executive Summary

Machine Learning (ML) is a subfield of Artificial Intelligence that focuses on building systems capable of learning patterns from data instead of being explicitly programmed. Arthur Samuel defined machine learning in 1959 as the field of study that gives computers the ability to learn without being explicitly programmed.

Machine learning powers many systems used daily, including web search ranking, recommendation systems, speech recognition, and image recognition. According to Andrew Ng’s introduction, machine learning is transforming industries such as healthcare, manufacturing, finance, transportation, and retail.

This lecture introduces the two primary categories of machine learning:

* **Supervised Learning** — learning from labeled data
* **Unsupervised Learning** — discovering patterns in unlabeled data

Understanding these paradigms is essential for building intelligent systems capable of making predictions, recognizing patterns, and supporting automated decision-making.

# 2. Supervised Learning

The algorithm learns a function:

# 3. Regression

Regression predicts **continuous numerical values**.

Example from the lecture: Predicting **house prices based on house size**.

---

# Regression Visualization



```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

house_size = np.random.randint(500,2500,50)
price = house_size*0.15 + np.random.randn(50)*20

plt.scatter(house_size,price)

coef = np.polyfit(house_size,price,1)
poly = np.poly1d(coef)

x = np.linspace(500,2500,100)
plt.plot(x,poly(x))

plt.xlabel("House Size (sq ft)")
plt.ylabel("Price ($1000s)")
plt.title("Regression Example: Housing Prices")

plt.show()
```

# 4. Classification

Classification predicts **categories or classes** rather than numbers.

Example:

Breast cancer detection

* Benign (0)
* Malignant (1)

This example appears in the lecture slides.

---

# Classification Visualization


```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X,y = make_blobs(n_samples=200,centers=2,random_state=42)

plt.scatter(X[:,0],X[:,1],c=y)

plt.title("Classification Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```

# 5. Unsupervised Learning

Unlike supervised learning, **unsupervised learning uses unlabeled data**.

The goal is to find patterns or structures within the data.

---

# Types of Unsupervised Learning

### 1. Clustering

Grouping similar data points.

Examples:

* Google News article grouping
* Customer segmentation
* DNA analysis

---

# Clustering Visualization



```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X,_ = make_blobs(n_samples=300,centers=3,random_state=42)

kmeans = KMeans(n_clusters=3)
labels = kmeans.fit_predict(X)

plt.scatter(X[:,0],X[:,1],c=labels)

plt.title("Clustering Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```

# Key Takeaways

* Machine learning enables computers to learn patterns from data.
* The two major ML paradigms are **supervised** and **unsupervised learning**.
* Supervised learning learns mappings from input (X) to output (Y).
* Two main supervised tasks:

  * Regression (predict numbers)
  * Classification (predict categories)
* Unsupervised learning finds structure in unlabeled data.
* Common unsupervised techniques include clustering, anomaly detection, and dimensionality reduction.
