## 🧠 Executive Summary

Machine Learning (ML) is a branch of Artificial Intelligence that enables computers to learn patterns from data rather than being explicitly programmed with rules. Instead of writing instructions for every possible situation, ML algorithms learn from examples and improve their performance as they process more data. This ability allows machines to solve complex problems such as recognizing speech, recommending movies, filtering spam emails, and ranking web search results.

The lectures introduce the basic definition of machine learning and explains why it has become so important in modern technology. Many real-world problems are too complex to solve with traditional programming because writing explicit rules for them is difficult or impossible. Machine learning solves this by enabling computers to learn patterns directly from data.

The lectures also introduce the **two fundamental types of machine learning: supervised learning and unsupervised learning**. Supervised learning uses labeled datasets to train models, while unsupervised learning works with unlabeled data and discovers hidden structures or patterns. These approaches form the foundation of most machine learning applications used today.

# 📚 Key Concepts & Glossary

**Machine Learning:**
A field of study that enables computers to learn from data without being explicitly programmed.

**Artificial Intelligence (AI):**
The broader field focused on building machines capable of performing tasks that normally require human intelligence.

**Artificial General Intelligence (AGI):**
A hypothetical form of AI that could perform any intellectual task that a human can.

**Learning Algorithm:**
A computational method that improves performance on a task by identifying patterns in data.

**Input (X):**
The data fed into a machine learning model.

**Output (Y):**
The predicted value or label produced by the algorithm.

**Supervised Learning:**
A type of machine learning where algorithms learn from labeled datasets containing both inputs and correct outputs.

**Unsupervised Learning:**
A type of machine learning where algorithms analyze unlabeled data to discover patterns or groupings.

**Regression:**
A supervised learning task that predicts continuous numerical values.

**Classification:**
A supervised learning task that predicts discrete categories.

**Clustering:**
An unsupervised learning technique that groups similar data points together.

**Anomaly Detection:**
A technique used to detect unusual patterns or outliers in data.

**Dimensionality Reduction:**
A process that reduces the number of features in a dataset while preserving important information.
