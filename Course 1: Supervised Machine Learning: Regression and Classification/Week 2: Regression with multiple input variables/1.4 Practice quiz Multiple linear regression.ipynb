## Question 1

**Question:**
True/False? To make gradient descent converge about twice as fast, a technique that almost always works is to double the learning rate *alpha*.

**Options:**
- True
- False

**Correct Answer:**
**False**

**Explanation:**
This statement is **False**. In **Gradient Descent**, the learning rate ($\alpha$) controls the size of the steps we take toward the minimum of the cost function.
1.  **Risk of Divergence:** If you double the learning rate, you risk making the steps too large. Instead of converging (settling into the minimum), the algorithm might overshoot the minimum completely and bounce back and forth, or even diverge (move further away from the solution).
2.  **No Linear Relationship:** There is no guarantee that doubling the learning rate will double the speed. Convergence depends on the shape of the cost function curve.
3.  **Best Practice:** To choose a good learning rate, practitioners typically try a range of values (e.g., 0.001, 0.003, 0.01, 0.03, 0.1, etc.) and see which one allows the cost function to decrease smoothly without oscillating.

**Real-World Example:**
Imagine you are trying to find the lowest point in a valley while blindfolded (this is the optimization problem).
-   **Small Learning Rate:** You take tiny baby steps. You will definitely find the bottom, but it will take a very long time.
-   **Doubling the Learning Rate (Too Large):** You decide to take giant leaps to get there faster. However, your leap is so big that you jump right over the bottom of the valley and land on the hill on the other side. Now you are further away from your goal than before!

## Question 2

**Question:**
In the training set below, what is $x_4^{(3)}$?

| Size in feet² ($x_1$) | Number of bedrooms ($x_2$) | Number of floors ($x_3$) | Age of home in years ($x_4$) | Price ($) in $1000's ($y$) |
| :--- | :--- | :--- | :--- | :--- |
| 2104 | 5 | 1 | 45 | 460 |
| 1416 | 3 | 2 | 40 | 232 |
| **1534** | **3** | **2** | **30** | **315** |
| 852 | 2 | 1 | 36 | 178 |

**Options:**
- 45
- 30
- 40
- 36

**Correct Answer:**
**30**

**Explanation:**
To solve this, we need to understand the notation used in **Multiple Linear Regression**:
-   $x$: Represents the input features.
-   Superscript $(i)$: Represents the **training example number** (the row number). So, $(3)$ refers to the **3rd row**.
-   Subscript $j$: Represents the **feature number** (the column number). So, $4$ refers to the **4th feature**.

Looking at the table:
1.  Go to the **3rd row** (ignoring the header). The values are: 1534, 3, 2, **30**, 315.
2.  Identify the **4th feature** ($x_4$). The columns are ordered: Size ($x_1$), Bedrooms ($x_2$), Floors ($x_3$), **Age ($x_4$)**.
3.  The value at the intersection of the 3rd row and 4th column is **30**.

**Real-World Example:**
Think of this like a spreadsheet (like Excel or Google Sheets).
-   The superscript $(3)$ is like saying "Row 3".
-   The subscript $4$ is like saying "Column D" (the 4th column).
-   You are simply looking up the specific data point for the "Age" of the "3rd house" listed in your database.

## Question 3

**Question:**
Which of the following are potential benefits of vectorization? Please choose the best option.

**Options:**
- It makes your code run faster
- It can make your code shorter
- It allows your code to run more easily on parallel compute hardware
- All of the above

**Correct Answer:**
**All of the above**

**Explanation:**
**Vectorization** is a technique (often using libraries like NumPy in Python) that allows you to perform operations on entire arrays of data at once, rather than using explicit `for` loops.
1.  **Faster Execution:** Vectorized code uses optimized low-level libraries (written in C or Fortran) that are much faster than standard Python loops.
2.  **Shorter Code:** You can replace 3-4 lines of loop code with a single line (e.g., `z = np.dot(w, x)` instead of a loop summing products). This makes code easier to read and maintain.
3.  **Parallel Hardware:** Modern CPUs and GPUs are designed to perform "Single Instruction, Multiple Data" (SIMD) operations. Vectorization allows the computer to multiply many numbers simultaneously using different cores, rather than doing them one by one sequentially.

**Real-World Example:**
Imagine you are a cashier scanning groceries.
-   **Non-Vectorized (Loop):** You scan one item, beep, put it in the bag. Then scan the next item, beep, bag it. You do this strictly one at a time.
-   **Vectorized:** Imagine a machine where you can slide a whole tray of 20 items through a scanner at once, and it instantly knows the total price of all of them simultaneously. This is how vectorization utilizes hardware to save time.
