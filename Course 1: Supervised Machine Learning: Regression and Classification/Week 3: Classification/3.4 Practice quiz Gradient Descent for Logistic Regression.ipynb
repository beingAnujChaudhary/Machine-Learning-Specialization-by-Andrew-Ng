# Practice quiz: Gradient Descent for Logistic Regression

## Question 1

**Which of the following two statements is a more accurate statement about gradient descent for logistic regression?**

image.png

- The update steps look like the update steps for linear regression, but the definition of  $f_{\vec{w}, b}(\vec{x}^{(i)})$  is different.
- The update steps are identical to the update steps for linear regression.

### Correct Answer

✅ **The update steps look like the update steps for linear regression, but the definition of  $f_{\vec{w}, b}(\vec{x}^{(i)})$  is different.**

### Explanation of Correct Answer

The first statement is accurate because:

1. **Same Update Rule Structure**: Both linear regression and logistic regression use the same gradient descent update pattern:

   $$
   w_j = w_j - \alpha \left[ \frac{1}{m} \sum_{i=1}^m (f_{\vec{w}, b}(\vec{x}^{(i)}) - y^{(i)})x_j^{(i)} \right]
   $$

   $$
   b = b - \alpha \left[ \frac{1}{m} \sum_{i=1}^m (f_{\vec{w}, b}(\vec{x}^{(i)}) - y^{(i)}) \right]
   $$

2. **Different Function Definitions**: The key difference lies in what $ f_{\vec{w}, b}(\vec{x}) $ represents:

   - **For linear regression**: $ f_{\vec{w}, b}(\vec{x}) = \vec{w} \cdot \vec{x} + b $ (a linear function)
   - **For logistic regression**: $ f_{\vec{w}, b}(\vec{x}) = \frac{1}{1 + e^{-(\vec{w} \cdot \vec{x} + b)}} $ (the sigmoid function)

3. **Why This Matters**: The sigmoid function squashes the output to a value between 0 and 1, which can be interpreted as a probability for classification tasks.

### Explanation of Incorrect Option

❌ **"The update steps are identical to the update steps for linear regression"**

This statement is incorrect because:

- While the **form** of the update equations appears identical, the **meaning** of $ f_{\vec{w}, b}(\vec{x}^{(i)}) $ is fundamentally different
- Linear regression uses a linear function that can output any real number
- Logistic regression uses the sigmoid function that outputs values between 0 and 1
- This difference reflects the different purposes: linear regression for prediction (regression) and logistic regression for classification

### Real-World Examples

#### Linear Regression Example
**Predicting house prices**: A real estate company wants to predict the selling price of a house based on features like square footage, number of bedrooms, and location. The output is a continuous value (e.g., $350,000) – any price is possible.

#### Logistic Regression Example
**Email spam detection**: An email service needs to classify incoming emails as either "spam" or "not spam." The algorithm outputs a probability between 0 and 1 (e.g., 0.85 probability of being spam), which is then thresholded to make a binary decision.

### Summary Table

| Aspect | Linear Regression | Logistic Regression |
|--------|-------------------|---------------------|
| **Purpose** | Predict continuous values | Predict binary outcomes |
| **Function $ f_{\vec{w}, b}(\vec{x}) $** | $ \vec{w} \cdot \vec{x} + b $ | $ \frac{1}{1 + e^{-(\vec{w} \cdot \vec{x} + b)}} $ |
| **Output Range** | $(-\infty, +\infty)$ | $(0, 1)$ |
| **Interpretation** | Direct prediction | Probability of belonging to class 1 |
| **Example** | House price prediction | Spam detection |

### Key Takeaway

The similarity in update rules is elegant – it shows that gradient descent is a general optimization algorithm that can be applied to different models. However, the models themselves (through their different $ f $ functions) are designed for fundamentally different tasks.
