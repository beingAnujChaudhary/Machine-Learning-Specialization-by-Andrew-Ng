# 📓 Practice quiz: The problem of overfitting


---

## Question 1: Methods to Address Overfitting

**Which of the following can address overfitting?**

- [ ] Collect more training data
- [ ] Select a subset of the more relevant features
- [ ] Apply regularization
- [ ] Remove a random set of training examples

---

### ✅ Correct Answer

**All options EXCEPT "Remove a random set of training examples" are correct methods to address overfitting.**

The correct answers are:
- **Collect more training data**
- **Select a subset of the more relevant features**
- **Apply regularization**

---

### 📚 Detailed Explanation

#### Why the First Three Options Work

| **Method** | **How It Addresses Overfitting** | **Mechanism** |
|------------|-----------------------------------|---------------|
| **Collect more training data** | Reduces overfitting by providing more examples of the underlying pattern, making it harder for the model to simply memorize noise | More data helps the model distinguish between signal (true pattern) and noise (random fluctuations) |
| **Select a subset of relevant features** | Reduces model complexity by eliminating features that may be causing the model to fit noise | Fewer features = simpler model = less opportunity to overfit |
| **Apply regularization** | Penalizes large parameter values, forcing the model to keep all features but reduce their impact | Adds penalty term $\frac{\lambda}{2m}\sum_{j=1}^{n}w_j^2$ to cost function, shrinking parameters |

#### Why "Remove a Random Set of Training Examples" is Incorrect

Removing training examples randomly **reduces** your training data, which typically **increases** overfitting rather than reducing it. With fewer examples, the model has less information about the true underlying pattern and is more likely to memorize the remaining examples.

**Think of it this way**: If you're studying for an exam and randomly throw away half your practice questions, you'll have fewer examples to learn from and are more likely to memorize the remaining ones rather than understanding the concepts.

---

### 🌍 Real-World Examples

#### Example 1: Collect More Training Data
**Scenario**: A self-driving car company developing a pedestrian detection system.

- **Initial model**: Trained on 10,000 images, performs well on training data but fails on new scenarios
- **Solution**: Collect 100,000 more images from diverse conditions (night, rain, different cities)
- **Result**: Model learns general features of pedestrians rather than memorizing specific training images

#### Example 2: Feature Selection
**Scenario**: A real estate company predicting house prices with 200 features.

- **Problematic features**: "Distance to nearest coffee shop" and "Number of trees on street" may cause overfitting
- **Solution**: Select only the most relevant features (size, bedrooms, location, age)
- **Result**: Simpler model that generalizes better to new houses

#### Example 3: Regularization
**Scenario**: A credit card company building a fraud detection model with 50 transaction features.

- **Problem**: Complex model catches all training fraud cases but misses new fraud patterns
- **Solution**: Apply L2 regularization with λ = 0.1 to shrink unimportant feature weights
- **Result**: Model focuses on truly predictive patterns rather than noise

## Question 2: Interpreting Model Performance

You fit logistic regression with polynomial features to a dataset, and your model creates a highly complex, wiggly decision boundary that perfectly classifies every training example.

image.png

**What would you conclude? (Pick one)**

- [ ] The model has high bias (underfit). Thus, adding data is likely to help
- [x] **The model has high variance (overfit). Thus, adding data is likely to help**
- [ ] The model has high bias (underfit). Thus, adding data is, by itself, unlikely to help much
- [ ] The model has high variance (overfit). Thus, adding data is, by itself, unlikely to help much

---

### ✅ Correct Answer

**The model has high variance (overfit). Thus, adding data is likely to help.**

---

### 📚 Detailed Explanation

#### Understanding the Scenario

From the lecture, when a logistic regression model with high-order polynomial features creates an overly complex decision boundary that perfectly classifies every training example but looks "wiggly" and unnatural, this is a classic sign of **overfitting (high variance)**.

#### Why Adding Data Will Help in This Case

Unlike the previous misconception, **adding more training data IS likely to help with overfitting**. Here's why:

1. **More data provides more constraints** - With more examples, the model cannot simply memorize; it must find patterns that work across all examples
2. **The decision boundary smooths out** - As more data points are added, the overly complex boundary becomes unsustainable
3. **Signal emerges from noise** - With sufficient data, the true underlying pattern becomes more apparent, and the model learns to ignore noise

#### The Learning Curve Effect

As you add more training data:
- **Training error** typically increases slightly (harder to fit all examples perfectly)
- **Test/validation error** typically decreases (better generalization)
- The gap between training and test error narrows

#### When Would Adding Data NOT Help?

Adding data is unlikely to help only when:
1. **The model has high bias (underfitting)** - The model is too simple to capture the pattern, regardless of how much data you have
2. **The data is of poor quality** - More bad data doesn't help
3. **You've reached the point of diminishing returns** - After a certain point, additional data provides minimal improvement

---

### 🌍 Real-World Example

**Scenario**: A handwriting recognition system for postal addresses.

- **Initial model**: Trained on 1,000 handwriting samples, creates overly complex boundaries that perfectly classify all training samples but fails on new handwriting styles
- **Diagnosis**: High variance (overfitting)
- **Solution**: Collect 100,000 more handwriting samples from diverse writers
- **Result**: With more examples, the model learns general handwriting patterns rather than memorizing specific writing styles. The decision boundaries become smoother and more generalizable.
- **Final performance**: Test accuracy improves from 75% to 94%

## Question 3: Effect of Increasing Regularization

### Regularization

image.png

$$\min_{\vec{w}, b} J(\vec{w}, b) = \min_{\vec{w}, b} \left[ \frac{1}{2m} \sum_{i=1}^m \left( f_{\vec{w}, b}(\vec{x}^{(i)}) - y^{(i)} \right)^2 + \frac{\lambda}{2m} \sum_{j=1}^n w_j^2 \right]$$

Suppose you have a regularized linear regression model. If you increase the regularization parameter $\lambda$, what do you expect to happen to the parameters $w_1, w_2, \dots, w_n$?

- [x] **This will reduce the size of the parameters $w_1, w_2, \dots, w_n$**
- [ ] This will increase the size of the parameters $w_1, w_2, \dots, w_n$

---

### ✅ Correct Answer

**Increasing the regularization parameter $\lambda$ will reduce the size of the parameters $w_1, w_2, \dots, w_n$.**

---

### 📚 Detailed Explanation

#### Understanding the Regularization Term

The regularization term in the cost function is:

$$\frac{\lambda}{2m} \sum_{j=1}^n w_j^2$$

Where:
- $\lambda$ = regularization parameter (controls penalty strength)
- $m$ = number of training examples
- $w_j$ = model parameters (weights)
- $n$ = number of features

#### How Increasing $\lambda$ Affects Parameters

When you increase $\lambda$:

1. **The penalty for large parameters increases** - The term $\frac{\lambda}{2m} \sum w_j^2$ becomes larger for any given set of $w_j$ values

2. **Gradient descent responds** - In the gradient descent update rule:
   $$w_j := w_j - \alpha\left[\frac{1}{m}\sum_{i=1}^{m}(f_{\vec{w},b}(\vec{x}^{(i)}) - y^{(i)})x_j^{(i)} + \frac{\lambda}{m}w_j\right]$$

   The term $\frac{\lambda}{m}w_j$ becomes larger, forcing $w_j$ to shrink more on each iteration

3. **Parameters shrink toward zero** - To minimize the cost function, the algorithm must make $w_j$ smaller to reduce the penalty term

#### Mathematical Intuition

At optimal solution, increasing $\lambda$ creates a tradeoff:
- **Smaller $\lambda$** : Focus on fitting training data (potentially larger $w_j$)
- **Larger $\lambda$** : Focus on keeping $w_j$ small (potentially worse fit to training data)

This is why very large $\lambda$ leads to underfitting - all $w_j$ become nearly zero.

---

### 🌍 Real-World Example

**Scenario**: A streaming service predicting user ratings for movies.

- **Features**: 50 features including user demographics, viewing history, time of day, device type
- **Without regularization (λ = 0)** : Some features get very large weights (e.g., "watched at 3 AM" might get weight 5.2)
- **With light regularization (λ = 0.1)** : All feature weights shrink moderately
- **With heavy regularization (λ = 10)** : All feature weights become very small (near zero), model essentially predicts the average rating for all users
- **Optimal λ** : Balances between overfitting (large weights on noise) and underfitting (all weights near zero)

## 📊 Summary Table: Overfitting Solutions

| **Method** | **How It Works** | **When to Use** | **Limitations** |
|------------|------------------|-----------------|------------------|
| **Collect more data** | Provides more examples of true pattern | When you have access to more high-quality data | May be expensive or impossible to collect |
| **Feature selection** | Reduces model complexity | When you have many irrelevant features | May discard useful information |
| **Regularization** | Penalizes large parameters | When you want to keep all features but reduce overfitting | Requires tuning λ correctly |
| **Remove random examples** | ❌ NOT RECOMMENDED | Never - this makes overfitting worse | Reduces information available to model |

---

## 🔑 Key Takeaways

1. **Overfitting occurs when the model is too complex** for the amount of training data available
2. **Three proven solutions**: More data, fewer features, or regularization
3. **Regularization parameter λ controls the tradeoff** between fitting training data and keeping parameters small
4. **Increasing λ shrinks parameters** toward zero, simplifying the model
5. **Model diagnosis matters** - identify whether you have high bias (underfitting) or high variance (overfitting) before choosing a solution
6. **For severe overfitting**, reducing model complexity (regularization or feature selection) is often more effective than simply adding more data

   *Answer: You get standard linear regression with no regularization (potentially overfitting)*

   *Answer: All parameters become nearly zero, resulting in underfitting (model predicts constant value)*
