## Question 1

**Question:**
In this lecture series, "cost" and "loss" have distinct meanings. Which one applies to a single training example?

image.png

**Options:**
- Loss
- Cost
- Both Loss and Cost
- Neither Loss nor Cost

---

### Correct Answer
**Loss**

### Explanation of Correct Answer
In machine learning, specifically in supervised learning contexts like logistic regression or neural networks:

*   **Loss Function ($L$):** This measures the error for a **single training example**. It tells you how far off your model's prediction is from the actual label for just one specific data point. In the formula shown in the image, $L(f_{\vec{w},b}(\vec{x}^{(i)}), y^{(i)})$ represents the loss for the $i$-th example.
*   **Cost Function ($J$):** This is the average of the loss functions over the **entire training set**. It gives you a global measure of how well your model is performing overall. In the formula, $J(\vec{w}, b) = \frac{1}{m} \sum_{i=1}^{m} L(...)$, the summation and division by $m$ (the number of examples) indicate that this is the Cost.

Since the question asks what applies to a *single* training example, the correct term is **Loss**.

### Explanation of Incorrect Options
*   **Cost:** This is incorrect because the Cost function applies to the *entire dataset*, not a single example. It is the aggregate measure used to update the model's parameters during training.
*   **Both Loss and Cost:** This is incorrect because, as defined in the lecture series mentioned, they have distinct meanings based on the scope (single example vs. whole dataset).
*   **Neither Loss nor Cost:** This is incorrect because "Loss" is the standard term for the error metric of a single example.

### Real-World Example
Imagine you are training a model to predict house prices (**Regression**).
*   **Loss:** You look at one specific house. The model predicts USD 300,000, but it actually sold for USD 320,000. The difference (error) for this single house is the **Loss**.
*   **Cost:** You look at all 1,000 houses in your dataset. You calculate the average error across all of them. This average is the **Cost**.

## Question 2

**Question:**
For the simplified loss function, if the label $y^{(i)} = 0$, then what does this expression simplify to?

image.png

**Options:**
- $-\log(1 - f_{\vec{w},b}(\mathbf{x}^{(i)}))$
- $\log(f_{\vec{w},b}(\mathbf{x}^{(i)}))$
- $\log(1 - f_{\vec{w},b}(\mathbf{x}^{(i)})) + \log(1 - f_{\vec{w},b}(\mathbf{x}^{(i)}))$
- $-\log(1 - f_{\vec{w},b}(\mathbf{x}^{(i)})) - \log(1 - f_{\vec{w},b}(\mathbf{x}^{(i)}))$

---

### Correct Answer
**$-\log(1 - f_{\vec{w},b}(\mathbf{x}^{(i)}))$** (The first option)

### Explanation of Correct Answer
The question provides the combined formula for binary cross-entropy loss:
$$L(f_{\vec{w},b}(\vec{x}^{(i)}), y^{(i)}) = -y^{(i)}\log(f_{\vec{w},b}(\vec{x}^{(i)})) - (1-y^{(i)})\log(1 - f_{\vec{w},b}(\vec{x}^{(i)}))$$

We need to simplify this when the label **$y^{(i)} = 0$**. Let's substitute $0$ for $y^{(i)}$ in the equation:

1.  **First Term:** $- (0) \cdot \log(f_{\vec{w},b}(\vec{x}^{(i)}))$
    *   Anything multiplied by 0 is 0. So, this entire term becomes **0**.

2.  **Second Term:** $- (1 - 0) \cdot \log(1 - f_{\vec{w},b}(\vec{x}^{(i)}))$
    *   $(1 - 0)$ simplifies to $1$.
    *   So, this term becomes $-1 \cdot \log(1 - f_{\vec{w},b}(\vec{x}^{(i)}))$.
    *   Which simplifies to **$-\log(1 - f_{\vec{w},b}(\vec{x}^{(i)}))$**.

Combining these, the total loss is just the second term: **$-\log(1 - f_{\vec{w},b}(\vec{x}^{(i)}))$**.

Alternatively, you can look at the piecewise definition provided at the top of Question 2. It explicitly states:
*   if $y^{(i)} = 0$, then $L = -\log(1 - f_{\vec{w},b}(\vec{x}^{(i)}))$.

### Explanation of Incorrect Options
*   **$\log(f_{\vec{w},b}(\mathbf{x}^{(i)}))$:** This would be part of the loss if $y^{(i)} = 1$ (and without the negative sign, which would make it a maximization problem rather than minimization). It is incorrect for $y=0$.
*   **$\log(1 - f...) + \log(1 - f...)$:** This adds two log terms together, which doesn't follow the algebraic simplification of the formula.
*   **$-\log(1 - f...) - \log(1 - f...)$:** This subtracts the term twice, which is mathematically incorrect based on the substitution.

### Real-World Example
Imagine a **Spam Detection** system (**Classification**).
*   $y=1$ means "Spam".
*   $y=0$ means "Not Spam" (Ham).
*   $f(x)$ is the probability the model thinks the email is Spam.

If an email is actually **Not Spam ($y=0$)**:
We want the model to predict a low probability of it being spam.
*   If the model predicts $f(x) = 0.9$ (thinks it's spam), then $(1 - 0.9) = 0.1$. The loss is $-\log(0.1)$, which is a large positive number (high penalty).
*   If the model predicts $f(x) = 0.1$ (correctly thinks it's not spam), then $(1 - 0.1) = 0.9$. The loss is $-\log(0.9)$, which is a small number (low penalty).

The formula $-\log(1 - f(x))$ effectively penalizes the model when it assigns a high probability of "Spam" to an email that is actually "Not Spam".
