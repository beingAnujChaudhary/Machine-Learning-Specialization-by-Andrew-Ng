# Practice quiz: Activation Functions

## Question 1

image.png

**For the task of predicting housing prices, which activation functions could you choose for the output layer? Choose the 2 options that apply.**

- [ ] ReLU
- [ ] linear
- [ ] Sigmoid

### Correct Answer Identification
The two correct options are:
- ✅ **linear** activation function
- ✅ **ReLU** activation function

### Explanation of Correct Answers

#### Linear Activation Function
The linear activation function (also called "no activation") is appropriate for **regression problems** where the target variable can take any real value (positive or negative). Housing prices are continuous values that can range from very low to very high, making linear activation a natural choice.

*Mathematical form:* $g(z) = z$

#### ReLU Activation Function
ReLU (Rectified Linear Unit) works well for housing price prediction because:
- Housing prices are **always non-negative** (you can't have a negative house price)
- ReLU outputs zero for negative inputs and the input value for positive inputs: $g(z) = \max(0, z)$
- This matches the real-world constraint that house prices ≥ $0

### Explanation of Incorrect Options

#### ❌ Sigmoid
The sigmoid activation function is **incorrect** for housing price prediction because:
- It outputs values strictly between 0 and 1: $g(z) = \frac{1}{1+e^{-z}}$
- Housing prices can be much larger than 1 (e.g., $500,000)
- Sigmoid is designed for **binary classification** problems where we want probabilities

### Real-World Examples

**Regression (Housing Prices):**
- Predicting the sale price of a home based on features like square footage, number of bedrooms, location, and age
- The model outputs a continuous dollar amount (e.g., $350,000)

**Binary Classification (where Sigmoid belongs):**
- **Email spam detection:** Classifying emails as either "spam" (1) or "not spam" (0)
- **Medical diagnosis:** Determining if a tumor is malignant (1) or benign (0)

## Question 2

image.png

**Which of the following activation functions is the most common choice for the hidden layers of a neural network?**

- [ ] ReLU (rectified linear unit)
- [ ] Sigmoid
- [ ] Linear
- [ ] Most hidden layers do not use any activation function

### Correct Answer Identification
- ✅ **ReLU (rectified linear unit)**

### Explanation of Correct Answer

ReLU has become the **default choice** for hidden layers in modern neural networks for several compelling reasons:

1. **Computational Efficiency:** ReLU is simple to compute ($\max(0,z)$), making training faster than sigmoid which requires exponential calculations

2. **Mitigates Vanishing Gradient Problem:** Unlike sigmoid, which squeezes gradients into a small range (0 to 0.25), ReLU maintains a constant gradient of 1 for positive inputs, allowing effective learning in deep networks

3. **Sparse Activation:** ReLU sets negative inputs to zero, creating sparse representations where only relevant neurons fire - this can improve learning and reduce overfitting

4. **Biological Plausibility:** ReLU's "on/off" behavior (firing only for positive inputs) loosely mimics how biological neurons work

### Explanation of Incorrect Options

#### ❌ Sigmoid
- Suffers from **vanishing gradient problem** - gradients become extremely small for large positive or negative inputs
- Outputs are not zero-centered, which can cause inefficient gradient updates
- Computationally more expensive due to exponential operations

#### ❌ Linear
- A network with only linear activations collapses to a **linear model**, regardless of depth
- Cannot learn complex, non-linear patterns in data
- Defeats the purpose of having multiple layers

#### ❌ No activation function
- Without activation functions, each layer performs only linear transformations
- The entire network becomes equivalent to a single linear transformation
- Cannot learn non-linear relationships

### Real-World Examples

**Why ReLU matters in practice:**
- **Image recognition:** A deep CNN using ReLU can learn hierarchical features - from edges to shapes to complex objects like faces or cars
- **Natural language processing:** Transformer models like BERT use ReLU-based activations to understand context and relationships between words

## Question 3

**True/False? A neural network with many layers but no activation function (in the hidden layers) is not effective; that's why we should instead use the linear activation function in every hidden layer.**

- [ ] False
- [ ] True

### Correct Answer Identification
- ✅ **False**

### Explanation

This statement is **false** because it contains two conceptual errors:

#### Error 1: The premise is correct but misleading
The first part is true: *"A neural network with many layers but no activation function is not effective"*

Without activation functions, each layer performs a linear transformation:
- Layer 1: $a^{[1]} = W^{[1]}x + b^{[1]}$
- Layer 2: $a^{[2]} = W^{[2]}a^{[1]} + b^{[2]}$
- Combined: $a^{[2]} = W^{[2]}(W^{[1]}x + b^{[1]}) + b^{[2]} = (W^{[2]}W^{[1]})x + (W^{[2]}b^{[1]} + b^{[2]})$

This is **still a linear function** of the input - stacking linear layers doesn't increase representational power!

#### Error 2: The proposed solution is wrong
The second part is false: *"that's why we should instead use the linear activation function in every hidden layer"*

Using linear activation functions in hidden layers still results in an overall linear model. The correct solution is to use **non-linear activation functions** (like ReLU) in hidden layers, which allows the network to learn complex patterns.

### Why Non-Linear Activations Are Essential

Think of a neural network like a recipe:
- **Linear transformations** are like individual ingredients
- **Non-linear activations** are like cooking processes that transform ingredients into something new
- Without cooking (non-linearity), you just have a pile of ingredients (linear combination) - not a complex dish!

### Correct Approach

For effective deep learning:
- **Hidden layers:** Use non-linear activations (typically ReLU)
- **Output layer:** Choose activation based on the task (linear/ReLU for regression, sigmoid for binary classification, softmax for multi-class)

### Real-World Example

**Why non-linearity matters:**
- **XOR problem:** A classic example where no linear model can solve the problem (points in opposite quadrants belong to same class)
- A neural network with one hidden layer using non-linear activation can easily learn the XOR pattern
- Without non-linear activations, the network fails regardless of depth

## Summary Table

| Layer Type | Recommended Activation | When to Use |
|------------|----------------------|-------------|
| Hidden Layers | **ReLU** | Default choice for most problems |
| Output Layer (Regression) | Linear or ReLU | Continuous values (ReLU if values ≥ 0) |
| Output Layer (Binary Classification) | Sigmoid | Probability output between 0 and 1 |
| Output Layer (Multi-class) | Softmax | Probability distribution across classes |

**Key Takeaway:** Non-linear activation functions in hidden layers are what give neural networks their power to learn complex patterns. Without them, depth doesn't matter - the network remains fundamentally linear.
