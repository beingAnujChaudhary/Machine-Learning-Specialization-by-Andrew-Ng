## Question 1

For the following code:

```python
model = Sequential([
    Dense(units=25, activation="sigmoid"),
    Dense(units=15, activation="sigmoid"),
    Dense(units=10, activation="sigmoid"),
    Dense(units=1, activation="sigmoid")
])
```

This code will define a neural network with how many layers?

a. 4

b. 5

c. 3

d. 25

### Correct Answer

✅ **Option a: 4**

### Explanation of Correct Answer

The code defines a neural network with **4 layers**. Here's why:

- In Keras/TensorFlow, each `Dense()` call creates one **layer** of neurons
- The `Sequential` container groups these layers together in sequence
- Looking at the code, we have four separate `Dense()` function calls:
  1. First layer: 25 neurons with sigmoid activation
  2. Second layer: 15 neurons with sigmoid activation  
  3. Third layer: 10 neurons with sigmoid activation
  4. Fourth layer: 1 neuron with sigmoid activation

Therefore, this neural network has **4 layers** total.

### Why Other Options Are Incorrect

- **Option b (5)**: This would be incorrect because there's no fifth layer. The `Sequential` container itself is not a layer - it's just a wrapper that holds the layers.

- **Option c (3)**: This undercounts the layers. Perhaps someone might mistakenly think the last layer doesn't count, but every `Dense()` call represents a distinct layer.

- **Option d (25)**: This refers to the number of neurons in the first layer, not the number of layers. It's a common beginner mistake to confuse the number of **neurons** with the number of **layers**.

### Key Concept: Neural Network Architecture

A neural network's architecture is defined by:
- **Number of layers** (depth)
- **Number of neurons per layer** (width)
- **Activation functions** used

In this example:
- **Depth**: 4 layers
- **Width**: 25 → 15 → 10 → 1 neurons
- **Activation**: Sigmoid for all layers

How do you define the second layer of a neural network that has 4 neurons and a sigmoid activation?

- [ ] Dense(units=4, activation='sigmoid')
- [ ] Dense(units=4)
- [ ] Dense(layer=2, units=4, activation='sigmoid')
- [ ] Dense(units=[4], activation=['sigmoid'])

### Correct Answer

✅ **Option a: Dense(units=4, activation='sigmoid')**

### Explanation of Correct Answer

The correct syntax for defining a dense layer in Keras/TensorFlow is:

```python
Dense(units=number_of_neurons, activation='activation_function')
```

For this specific case, we want:
- **4 neurons**: specified by `units=4`
- **Sigmoid activation**: specified by `activation='sigmoid'`

The question asks specifically about defining a **layer** - not about which layer number it is in the sequence.

### Why Other Options Are Incorrect

- **Option b (Dense(units=4))**: This creates a layer with 4 neurons but **no activation function**. Without specifying activation, the layer would perform only a linear transformation, which severely limits what the network can learn.

- **Option c (Dense(layer=2, units=4, activation='sigmoid'))**: This uses invalid syntax. The `Dense` class doesn't take a `layer` parameter - the layer order is determined by the sequence in the model, not by a parameter.

- **Option d (Dense(units=[4], activation=['sigmoid']))**: This uses incorrect syntax. Both `units` and `activation` should be single values, not lists, for a standard dense layer.

### Key Concept: Layer Definition Syntax

When defining a layer in Keras/TensorFlow, remember:
- `units` = number of neurons in the layer
- `activation` = activation function applied to the layer's output
- Each layer is independent - you don't specify "this is layer 2" explicitly
- The layer number is determined by its position in the `Sequential` list

### Feature vectors

| temperature (Celsius) | duration (minutes) | Good coffee? (1/0) |
|----------------------|-------------------|-------------------|
| 200.0                | 17.0              | 1                 |
| 425.0                | 18.5              | 0                 |

If the input features are temperature (in Celsius) and duration (in minutes), how do you write the code for the first feature vector x shown above?

- [ ] `x = np.array([[200.0, 17.0]])`
- [ ] `x = np.array([[200.0],[17.0]])`
- [ ] `x = np.array([['200.0', '17.0']])`
- [ ] `x = np.array([[200.0 + 17.0]])`

### Correct Answer

✅ **Option a: `x = np.array([[200.0, 17.0]])`**

### Explanation of Correct Answer

The correct representation of a **single feature vector** with **two features** is a 2D NumPy array with:
- **1 row** (representing 1 example)
- **2 columns** (representing 2 features)

Breaking down `np.array([[200.0, 17.0]])`:
- Outer brackets `[]` create the array
- Inner brackets `[200.0, 17.0]` represent one row with two values
- This creates a shape `(1, 2)` array - perfect for a single training example

### Why Other Options Are Incorrect

- **Option b (`np.array([[200.0],[17.0]])`)**: This creates a **column vector** with shape `(2, 1)`. While it contains the same numbers, the structure is wrong:
  - 2 rows, 1 column (instead of 1 row, 2 columns)
  - This would be interpreted as 2 examples, each with 1 feature

- **Option c (`np.array([['200.0', '17.0']])`)**: This uses **strings** instead of numbers:
  - Machine learning models require numerical inputs
  - Strings cannot be used in mathematical operations
  - This would cause a type error when training

- **Option d (`np.array([[200.0 + 17.0]])`)**: This performs the addition **before** creating the array:
  - Results in `np.array([[217.0]])`
  - Loses the individual feature information
  - Shape becomes `(1, 1)` instead of `(1, 2)`

### Key Concept: Data Representation in NumPy

When preparing data for neural networks:

1. **Each row = one training example**
2. **Each column = one feature**
3. **Data type must be numeric** (float or integer)

For a dataset with:
- **m** training examples
- **n** features

The input array should have shape **(m, n)**

In this case:
- m = 1 (first example only)
- n = 2 (temperature and duration)
- Shape = (1, 2)

### Real-World Example

Think of it like a spreadsheet:
- Each row is a different coffee brewing attempt
- Each column is a measurement (temperature, time)
- The array structure preserves this organization

## Key Formulas

When working with neural network layers, the transformation performed by a dense layer is:

$$a = g(Wx + b)$$

Where:
- **x** = input to the layer (feature vector)
- **W** = weight matrix
- **b** = bias vector
- **g** = activation function (sigmoid, ReLU, etc.)
- **a** = output of the layer (activation values)

For a layer with:
- **n** input features
- **m** neurons

The dimensions are:
- **W** is m × n matrix
- **b** is m-dimensional vector
- **x** is n-dimensional vector
- **a** is m-dimensional vector
