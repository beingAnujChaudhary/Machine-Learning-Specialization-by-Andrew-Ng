## Question 1: Binary Cross Entropy Use Case

image.png

Here is some code that you saw in the lecture:

```
model.compile(loss=BinaryCrossentropy())

```
**Question:** For which type of task would you use the binary cross entropy loss function?

**Options:**
- ✅ binary classification (classification with exactly 2 classes)
- ❌ BinaryCrossentropy() should not be used for any task.
- ❌ A classification task that has 3 or more classes (categories)
- ❌ regression tasks (tasks that predict a number)

### Correct Answer
✅ **binary classification (classification with exactly 2 classes)**

### Why This is Correct

**Binary Cross Entropy** is specifically designed for **binary classification problems** where:
- The output has **exactly two possible classes**
- The final layer has **1 neuron** with **sigmoid activation**
- The output represents the probability of belonging to the positive class (between 0 and 1)

The formula for binary cross entropy is:

$$L = -[y \log(\hat{y}) + (1-y) \log(1-\hat{y})]$$

Where:
- $y$ = true label (0 or 1)
- $\hat{y}$ = predicted probability (between 0 and 1)
- $\log$ = natural logarithm

This loss function:
- **Penalizes confidently wrong predictions heavily**
- **Rewards confident correct predictions**
- Works perfectly when you need to distinguish between two categories

### Why Other Options Are Incorrect

**Option 2 (should not be used):** Binary cross entropy is actually one of the most commonly used loss functions in deep learning!

**Option 3 (3+ classes):** For multi-class classification, you need **Categorical Cross Entropy** (if labels are one-hot encoded) or **Sparse Categorical Cross Entropy** (if labels are integers).

**Option 4 (regression):** Regression tasks need different loss functions like **Mean Squared Error (MSE)** or **Mean Absolute Error (MAE)**.

### Real-World Examples

**Binary Classification Examples:**
- **Spam Detection:** Is an email spam (1) or not spam (0)?
- **Medical Diagnosis:** Does a patient have a disease (1) or not (0)?
- **Customer Churn:** Will a customer cancel their subscription (1) or stay (0)?
- **Credit Card Fraud:** Is a transaction fraudulent (1) or legitimate (0)?

**Non-Examples:**
- **Multi-class:** Identifying handwritten digits (0-9) → use categorical cross entropy
- **Regression:** Predicting house prices → use mean squared error

## Question 2: Network Parameter Updates

image.png

Here is code that you saw in the lecture:

```
model = Sequential([

Dense(units=25, activation='sigmoid’),

Dense(units=15, activation='sigmoid’),

Dense(units=1, activation='sigmoid’)

])



model.compile(loss=BinaryCrossentropy())

model.fit(X,y,epochs=100)
```

**Question:** Which line of code updates the network parameters in order to reduce the cost?

**Options:**
- ✅ `model.fit(X, y, epochs=100)`
- ❌ `model.compile(loss=BinaryCrossentropy())`
- ❌ `model = Sequential([...])`
- ❌ None of the above

### Correct Answer
✅ **`model.fit(X, y, epochs=100)`**

### Why This is Correct

The training process in TensorFlow follows these steps:

1. **`Sequential([...])`** → **Architecture definition**
   - Creates the network structure
   - Initializes parameters (weights and biases) randomly

2. **`compile(loss=BinaryCrossentropy())`** → **Configuration**
   - Sets up the loss function
   - Configures the optimizer (Adam by default)
   - Prepares metrics to track

3. **`fit(X, y, epochs=100)`** → **Actual training!**
   - Performs forward propagation
   - Calculates the loss
   - Does backpropagation
   - **Updates parameters** using gradient descent
   - Repeats for specified number of epochs

### The Training Loop Visualized

```
Epoch 1:
    Forward prop → Calculate loss → Backward prop → Update weights
    ↓
Epoch 2:
    Forward prop → Calculate loss → Backward prop → Update weights
    ↓
... (repeats 100 times)
```

### Why Other Options Are Incorrect

**`model.compile()`** only configures how training will happen - it doesn't actually train anything. Think of it like setting up your workout plan vs. actually exercising.

**`model = Sequential([])`** just defines the architecture and initializes random parameters. No learning happens here.

### Real-World Analogy

Think of training a neural network like learning to play golf:

| Step | Code | Analogy |
|------|------|---------|
| `Sequential([...])` | Architecture definition | Getting your clubs ready |
| `compile(...)` | Configuration | Deciding you'll count strokes (loss) and choosing your practice method (optimizer) |
| **`fit(...)`** | **Actual training** | **Actually swinging the club, seeing where the ball goes, and adjusting your swing** |

The `fit()` method is where the actual learning happens - each swing (forward pass) is evaluated (loss calculation), and you adjust your form (parameter updates) to improve (reduce loss).

### Key Takeaway

The `fit()` method is the workhorse of TensorFlow training - it's where the magic of gradient descent and parameter updates actually occurs!
