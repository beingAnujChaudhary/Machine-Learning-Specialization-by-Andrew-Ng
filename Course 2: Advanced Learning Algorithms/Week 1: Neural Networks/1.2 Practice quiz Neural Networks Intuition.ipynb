# Neural Networks Intuition - Practice Questions

## Question 1

**Which of these are terms used to refer to components of an artificial neural network? (hint: three of these are correct)**

image.png

- [ ] neurons
- [ ] layers
- [ ] activation function
- [ ] axon

---

### Correct Answer

The three correct options are:
- ✅ **neurons**
- ✅ **layers**
- ✅ **activation function**

---

### Explanation of Correct Answers

#### 1. Neurons ✓
**Why it's correct:** Artificial **neurons** (also called "units") are the fundamental building blocks of neural networks. Each neuron receives one or more inputs, performs a computation (typically a weighted sum followed by an activation function), and produces an output.

**Mathematical representation of a single neuron:**
$$a = g(\sum_{i=1}^{n} w_i x_i + b)$$

Where:
- $x_i$ = input features
- $w_i$ = weights (importance of each input)
- $b$ = bias term
- $g$ = activation function
- $a$ = activation output

**Real-world analogy:** Think of a neuron as a small decision-maker in a company. It receives information from multiple sources (inputs), weighs their importance (weights), and makes a decision (output) that gets passed to other decision-makers.

---

#### 2. Layers ✓
**Why it's correct:** **Layers** are groups of neurons organized together. Neural networks are structured as a sequence of layers:
- **Input layer**: Receives raw data
- **Hidden layers**: Process information and learn features
- **Output layer**: Produces final prediction

**Layer visualization:**
```
Input Layer    Hidden Layer    Output Layer
    ○─────○─────○
    ○─────○
    ○─────○─────○
    ○─────○
```

**Real-world example:** Like an assembly line in manufacturing:
- Input layer = Raw materials entering the factory
- Hidden layers = Different assembly stations, each adding complexity
- Output layer = Finished product

---

#### 3. Activation Function ✓
**Why it's correct:** An **activation function** introduces non-linearity into the network, allowing it to learn complex patterns. Without activation functions, neural networks would just be linear models regardless of their depth.

**Common activation functions:**

**Sigmoid (for binary classification):**
$$g(z) = \frac{1}{1 + e^{-z}}$$

**ReLU (Rectified Linear Unit, most common for hidden layers):**
$$g(z) = max(0, z)$$

**Tanh (hyperbolic tangent):**
$$g(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}}$$

**Real-world analogy:** An activation function is like a "gate" that decides whether a neuron should "fire" or not based on the strength of input signals.

---

### Explanation of Incorrect Option

#### 4. Axon ✗
**Why it's incorrect:** **Axon** is a biological term from neuroscience—it's the part of a biological neuron that transmits signals to other neurons. While neural networks were originally inspired by biology, modern artificial neural networks use different terminology:

| Biological Term | Artificial Neural Network Term |
|----------------|-------------------------------|
| Dendrites (inputs) | Input features / connections |
| Cell body (processing) | Neuron / unit |
| Axon (output) | Output / activation |
| Synapse (connection) | Weight |

**Key distinction:** Artificial neural networks use mathematical abstractions, not biological structures. We don't use terms like "axon," "dendrite," or "synapse" when describing artificial neural networks—we use terms like **weights**, **activations**, and **connections**.

---

### Real-World Examples of Neural Network Components

| Component | Real-World Example | How It Works |
|-----------|-------------------|--------------|
| **Neurons** | Spam detection system | Each neuron learns to detect specific patterns (e.g., suspicious words, sender reputation) |
| **Layers** | Face recognition app | Layer 1: Detects edges; Layer 2: Detects facial features; Layer 3: Identifies faces |
| **Activation** | Stock price prediction | Decides whether certain market signals are strong enough to influence prediction |

## Question 2

**True/False? Neural networks take inspiration from, but do not very accurately mimic, how neurons in a biological brain learn.**

- [ ] True
- [ ] False

---

### Correct Answer

✅ **True**

---

### Explanation

This statement is **true** for several important reasons:

#### Historical Context
Neural networks were indeed inspired by the brain in the 1950s and 1960s. Researchers observed that biological neurons receive signals, process them, and transmit outputs—and they created mathematical models to simulate this process.

#### The Reality Gap
However, modern artificial neural networks are **extremely simplified** compared to biological brains:

| Aspect | Biological Brain | Artificial Neural Network |
|--------|-----------------|--------------------------|
| **Scale** | ~86 billion neurons | Typically millions of parameters |
| **Connectivity** | Complex, dynamic, sparse | Usually dense (all-to-all) |
| **Signal type** | Chemical/electrical pulses | Floating-point numbers |
| **Learning** | Hebbian learning, neuroplasticity | Backpropagation + gradient descent |
| **Energy use** | ~20 watts | Often thousands of watts |
| **Structure** | 3D, complex topology | Layered, feedforward typically |

#### Andrew Ng's Perspective
As mentioned in the lecture: *"Even though today's neural networks have almost nothing to do with how the brain learns... those of us that do research in deep learning have shifted away from looking to biological motivation that much, but instead are just using engineering principles to figure out how to build algorithms that are more effective."*

---

### Common Misconception Clarified

**Why might someone think this is false?**
Some people believe neural networks accurately model the brain because of the terminology overlap (neurons, activation) and the original biological inspiration.

**The truth:** We still understand very little about how the brain actually works. Every few years, neuroscientists make fundamental breakthroughs about brain function—a sign that there's much more to discover. Attempting to blindly mimic what little we know wouldn't lead to effective AI.

---

### Real-World Examples

| Application | What Neural Networks Actually Do | What Brains Would Do |
|------------|----------------------------------|---------------------|
| **Image recognition** | Process pixel values through mathematical operations | Integrate visual context, attention, prior knowledge, emotions |
| **Language translation** | Map sequences to sequences using statistical patterns | Understand meaning, cultural context, speaker intent |
| **Game playing (AlphaGo)** | Evaluate board positions through learned heuristics | Strategic reasoning, intuition, creativity |

---

### Key Takeaway

The biological inspiration was a **starting point** for neural networks, but modern deep learning is driven by **engineering principles**, mathematical optimization, and empirical results—not neuroscience.

## Key Terminology Review

| Term | Definition | Neural Network Role |
|------|------------|---------------------|
| **Neuron** | Basic computational unit | Processes inputs, produces output |
| **Layer** | Group of neurons | Organizes computation hierarchically |
| **Activation function** | Non-linear transformation | Enables learning complex patterns |
| **Weight** | Connection strength | Determines input importance |
| **Bias** | Offset term | Allows activation threshold adjustment |
