## Question 1: Entropy as a measure of impurity

image.png

**Question:**
Recall that entropy was defined in lecture as $H(p_1) = -p_1 \log_2(p_1) - p_0 \log_2(p_0)$, where $p_1$ is the fraction of positive examples and $p_0$ the fraction of negative examples.

At a given node of a decision tree, 6 of 10 examples are cats and 4 of 10 are not cats. Which expression calculates the entropy $H(p_1)$ of this group of 10 animals?

**Options:**
- $-(0.6)\log_2(0.6) - (0.4)\log_2(0.4)$
- $-(0.6)\log_2(0.6) - (1 - 0.4)\log_2(1 - 0.4)$
- $(0.6)\log_2(0.6) + (0.4)\log_2(0.4)$
- $(0.6)\log_2(0.6) + (1 - 0.4)\log_2(1 - 0.4)$

### Correct Answer
**Option 1:** $-(0.6)\log_2(0.6) - (0.4)\log_2(0.4)$

### Explanation
**Entropy** is a measure of impurity or disorder in a dataset. In binary classification (like Cat vs. Not Cat), the formula is:
$$H(p_1) = -p_1 \log_2(p_1) - p_0 \log_2(p_0)$$

1.  **Identify $p_1$:** This is the fraction of positive examples (cats).
    $$p_1 = \frac{6}{10} = 0.6$$
2.  **Identify $p_0$:** This is the fraction of negative examples (not cats). Since probabilities sum to 1:
    $$p_0 = 1 - p_1 = 1 - 0.6 = 0.4$$
3.  **Substitute into the formula:**
    $$H(0.6) = -(0.6)\log_2(0.6) - (0.4)\log_2(0.4)$$

**Why the other options are incorrect:**
-   **Option 2:** This substitutes $(1-0.4)$ for $p_0$, which equals $0.6$. While mathematically $0.6$ is correct for $p_1$, the structure implies a confusion between $p_1$ and $p_0$ in the second term. More importantly, the standard form uses the calculated probabilities directly.
-   **Options 3 & 4:** These are missing the negative signs. Entropy must be positive, but $\log_2$ of a number between 0 and 1 is negative. The negative signs in the formula cancel this out to produce a positive entropy value. Without the negative signs, the result would be negative, which is impossible for entropy.

**Real-World Example:**
Imagine a bag of marbles.
-   **Low Entropy:** A bag with 10 red marbles and 0 blue marbles. It's very "pure." You know exactly what you will pull out.
-   **High Entropy:** A bag with 5 red marbles and 5 blue marbles. It's very "impure" or disordered. You have maximum uncertainty about what you will pull out.

## Question 2: Information Gain

image.png

**Question:**
Recall that information gain was defined as follows:
$$Information\ Gain = H(p_1^{root}) - \left(w^{left} H(p_1^{left}) + w^{right} H(p_1^{right})\right)$$

Before a split, the entropy of a group of 5 cats and 5 non-cats is $H(5/10)$.
After splitting on a particular feature, a group of 7 animals (4 of which are cats) has an entropy of $H(4/7)$. The other group of 3 animals (1 is a cat) and has an entropy of $H(1/3)$. What is the expression for information gain?

**Options:**
- $H(0.5) - \left(\frac{7}{10}H(4/7) + \frac{3}{10}H(1/3)\right)$
- $H(0.5) - (H(4/7) + H(1/3))$
- $H(0.5) - \left(\frac{4}{7} * H(4/7) + \frac{4}{7} * H(1/3)\right)$
- $H(0.5) - (7 * H(4/7) + 3 * H(1/3))$

### Correct Answer
**Option 1:** $H(0.5) - \left(\frac{7}{10}H(4/7) + \frac{3}{10}H(1/3)\right)$

### Explanation
**Information Gain** measures how much "uncertainty" is reduced after splitting the data. It is calculated as the Entropy of the parent node minus the **weighted average** entropy of the child nodes.

1.  **Parent Entropy ($H^{root}$):** The root has 5 cats and 5 non-cats (10 total).
    $$p_1 = 5/10 = 0.5$$
    So, the first term is $H(0.5)$.
2.  **Left Child Weight ($w^{left}$):** The left group has 7 animals out of the original 10.
    $$w^{left} = \frac{7}{10}$$
    Its entropy is given as $H(4/7)$.
3.  **Right Child Weight ($w^{right}$):** The right group has 3 animals out of the original 10.
    $$w^{right} = \frac{3}{10}$$
    Its entropy is given as $H(1/3)$.
4.  **Combine:**
    $$Gain = H(0.5) - \left( \frac{7}{10}H(4/7) + \frac{3}{10}H(1/3) \right)$$

**Why the other options are incorrect:**
-   **Option 2:** This simply subtracts the sum of the entropies without weighting them by the size of the groups. A large group with high entropy should penalize the split more than a tiny group with high entropy.
-   **Option 3:** The weights used ($\frac{4}{7}$) are incorrect. Weights must be based on the proportion of the *total* dataset ($\frac{7}{10}$ and $\frac{3}{10}$), not the internal probability of the node.
-   **Option 4:** This multiplies by the count (7 and 3) rather than the fraction/weight. This would result in a massive number that doesn't represent an average.

**Real-World Example:**
Imagine you are trying to guess a number between 1 and 100.
-   **Before splitting:** High uncertainty (High Entropy).
-   **Question:** "Is the number greater than 50?"
-   **After splitting:** You eliminate half the possibilities. Your uncertainty drops significantly. The amount of uncertainty you lost is the **Information Gain**.

## Question 3: One-hot encoding

image.png

**Question:**
To represent 3 possible values for the ear shape, you can define 3 features for ear shape: pointy ears, floppy ears, oval ears. For an animal whose ears are not pointy, not floppy, but are oval, how can you represent this information as a feature vector?

**Options:**
- [0, 0, 1]
- [1, 1, 0]
- [0, 1, 0]
- [1, 0, 0]

### Correct Answer
**Option 1:** [0, 0, 1]

### Explanation
**One-Hot Encoding** is a way to convert categorical data (like text labels) into numbers that machine learning models can understand. We create a binary vector (containing only 0s and 1s) where only one bit is "hot" (1) and the rest are "cold" (0).

The features are defined in this order:
1.  Pointy ears
2.  Floppy ears
3.  Oval ears

The animal has **Oval** ears.
-   Is it Pointy? No $\rightarrow$ 0
-   Is it Floppy? No $\rightarrow$ 0
-   Is it Oval? Yes $\rightarrow$ 1

Resulting Vector: **[0, 0, 1]**

**Why the other options are incorrect:**
-   **[1, 1, 0]:** This would imply the animal has BOTH pointy and floppy ears, which is physically impossible in this context (categories are mutually exclusive).
-   **[0, 1, 0]:** This represents "Floppy ears".
-   **[1, 0, 0]:** This represents "Pointy ears".

**Real-World Example:**
Think of a menu at a restaurant where you can choose only one side dish: Fries, Salad, or Soup.
-   If you order Fries: [1, 0, 0]
-   If you order Salad: [0, 1, 0]
-   If you order Soup: [0, 0, 1]
You cannot order [1, 1, 0] because that would mean you ordered both Fries and Salad as your single side choice.

## Question 4: Stopping Criteria

**Question:**
Which of these are commonly used criteria to decide to stop splitting? (Choose two.)

**Options:**
- When the tree has reached a maximum depth
- When the number of examples in a node is below a threshold
- When the information gain from additional splits is too large
- When a node is 50% one class and 50% another class (highest possible value of entropy)

### Correct Answers
**Option 1:** When the tree has reached a maximum depth
**Option 2:** When the number of examples in a node is below a threshold

### Explanation
To prevent **Overfitting** (where the model memorizes the training data too perfectly and fails on new data), we must stop the tree from growing indefinitely.

1.  **Maximum Depth:** We set a limit on how many layers deep the tree can grow. This prevents the tree from becoming too complex and specific to noise in the data.
2.  **Minimum Samples per Node:** If a node has very few examples (e.g., only 2 animals left), splitting it further might just be fitting to outliers rather than learning a general rule. We stop splitting if the group is too small.

**Why the other options are incorrect:**
-   **"Information gain... is too large":** This is incorrect. We *want* large information gain. Large gain means the split was very useful in separating the classes. We usually stop if the gain is too *small* (below a threshold).
-   **"Node is 50% one class...":** This is actually the state of **maximum impurity** (maximum entropy). This is the exact scenario where we *most* want to split! We split to reduce this 50/50 uncertainty. We stop when the node is pure (100% one class), not when it is mixed.

**Real-World Example:**
Imagine you are hiring employees based on a decision tree.
-   **Max Depth:** You decide to only ask 3 interview questions. If you ask 50 questions, you might hire someone who is great at answering trivia but bad at the actual job (Overfitting).
-   **Min Samples:** You decide not to make a hiring rule based on a group of only 1 person. "We only hire people named Steve who wear red hats" is a bad rule if it's based on just one previous employee.

## Question 5: Splitting on a continuous variable

image.png

**Question:**
For a continuous valued feature (such as weight of the animal), there are 10 animals in the dataset. According to the lecture, what is the recommended way to find the best split for that feature?

**Options:**
- Choose the 9 mid-points between the 10 examples as possible splits, and find the split that gives the highest information gain.
- Try every value spaced at regular intervals (e.g., 8, 8.5, 9, 9.5, 10, etc.) and find the split that gives the highest information gain.
- Use gradient descent to find the value of the split threshold that gives the highest information gain.
- Use a one-hot encoding to turn the feature into a discrete feature vector of 0's and 1's, then apply the algorithm we had discussed for discrete features.

### Correct Answer
**Option 1:** Choose the 9 mid-points between the 10 examples as possible splits, and find the split that gives the highest information gain.

### Explanation
When dealing with **continuous variables** (numbers that can take any value, like weight or height), there are infinite possible split points. We cannot check every single number (e.g., 8.1, 8.11, 8.111...).

The efficient algorithm is:
1.  Sort the examples by the continuous feature (e.g., weight).
2.  Look at the gaps between adjacent examples.
3.  The optimal split point will always lie somewhere between two examples that have different labels.
4.  Therefore, we only need to test the **midpoints** between the sorted values. If there are 10 examples, there are 9 gaps (midpoints) between them. We calculate Information Gain for each of these 9 splits and pick the best one.

**Why the other options are incorrect:**
-   **"Regular intervals":** This is inefficient and arbitrary. The best split might be at 8.3, which you would miss if you only checked 8.0 and 8.5.
-   **"Gradient descent":** Decision Trees are not differentiable functions (you can't take a derivative of a "if/else" split), so standard gradient descent doesn't apply to finding split thresholds in standard CART algorithms.
-   **"One-hot encoding":** You cannot one-hot encode a continuous variable directly because there are infinite unique values. You would end up with infinite features.

**Real-World Example:**
Imagine you want to split people into "Tall" and "Short" based on height. You have a list of 10 people sorted by height.
-   Person A is 160cm (Short)
-   Person B is 162cm (Short)
-   Person C is 175cm (Tall)
-   ...
The best cutoff isn't necessarily 170cm. It's likely somewhere between Person B (162) and Person C (175). The midpoint is $\frac{162+175}{2} = 168.5$cm. We test these midpoints to find the perfect cutoff.
