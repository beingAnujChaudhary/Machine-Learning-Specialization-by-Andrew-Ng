## Question 1: Choosing the Best Split

image.png

**Question:**  
Take a decision tree learning to classify between spam and non-spam email. There are 20 training examples at the root node, comprising 10 spam and 10 non-spam emails. If the algorithm can choose from among four features, resulting in four corresponding splits, which would it choose (i.e., which has highest purity)?

**Options:**
- [ ] Left split: 10 of 10 emails are spam. Right split: 0 of 10 emails are spam.
- [ ] Left split: 7 of 8 emails are spam. Right split: 3 of 12 emails are spam.
- [ ] Left split: 2 of 2 emails are spam. Right split: 8 of 18 emails are spam.
- [ ] Left split: 5 of 10 emails are spam. Right split: 5 of 10 emails are spam.

---

### Correct Answer

✅ **Option 1:** Left split: 10 of 10 emails are spam. Right split: 0 of 10 emails are spam.

---

### Explanation of the Correct Answer

#### The Core Concept: Impurity and Purity

When building a decision tree, the algorithm at each node asks: *"Which feature should I split on to best separate the classes?"* The goal is to maximize **purity** (or minimize **impurity**) in the child nodes after the split.

- **Purity** means that after the split, each child node contains mostly (or entirely) examples from a single class.
- **Impurity** means that after the split, the child nodes still have a mix of classes.

#### Why Option 1 is Best

Let's examine the purity of each split:

**Option 1:**  
- Left node: 10/10 spam → **100% pure** (all spam)  
- Right node: 0/10 spam → **100% pure** (all non-spam)  

This split perfectly separates the classes. There is no uncertainty left in either child node. This is the ideal scenario for a decision tree.

**Option 2:**  
- Left node: 7/8 spam → 87.5% spam  
- Right node: 3/12 spam → 25% spam  

While better than random, there is still some mixing (impurity) in both nodes.

**Option 3:**  
- Left node: 2/2 spam → 100% pure  
- Right node: 8/18 spam → ~44.4% spam  

Although the left node is pure, the right node is highly impure (almost a 50-50 mix).

**Option 4:**  
- Left node: 5/10 spam → 50% spam  
- Right node: 5/10 spam → 50% spam  

This split does nothing to separate the classes. Both child nodes are as impure as the parent node.

#### Visual Comparison

| Split | Left Node Purity | Right Node Purity | Overall Separation |
|-------|------------------|-------------------|-------------------|
| Option 1 | 100% | 100% | Perfect |
| Option 2 | 87.5% | 75% (non-spam) | Good |
| Option 3 | 100% | 44.4% | Poor (right node impure) |
| Option 4 | 50% | 50% | No separation |

---

### Why the Other Options Are Incorrect

- **Option 2:** This split does improve purity compared to the root (which was 50-50), but it's not as good as Option 1. There's still uncertainty in both nodes.
  
- **Option 3:** While one node is perfectly pure, the other node is almost as impure as the starting point. Decision trees prefer splits where **all** resulting nodes are pure, not just one.
  
- **Option 4:** This split provides no benefit at all. The children have the same class distribution as the parent, so the split is useless for classification.

---

### Real-World Example

**Concept: Classification with Decision Trees**

Imagine you're a doctor trying to diagnose whether a patient has a certain disease based on symptoms.

- **Root node:** 20 patients, 10 with disease, 10 without.
- You consider different tests (features):
  - **Test 1 (Option 1):** Splits patients perfectly. Those with positive result all have disease; those with negative result all are healthy.
  - **Test 2 (Option 2):** Most with positive result have disease, but some false positives and false negatives remain.
  - **Test 3 (Option 3):** Positive result is highly reliable, but negative result still includes many with disease.
  - **Test 4 (Option 4):** Test result gives no useful information at all.

Clearly, Test 1 is the best diagnostic tool. Similarly, decision trees choose the feature that best separates the classes.

## Question 2: Traversing a Decision Tree

image.png

**Question:**  
Based on the decision tree shown in the lecture, if an animal has floppy ears, a round face shape and has whiskers, does the model predict that it's a cat or not a cat?

**Options:**
- [ ] cat
- [ ] Not a cat

---

### Correct Answer

✅ **cat**

---

### Explanation of the Correct Answer

#### How to Traverse a Decision Tree

To make a prediction with a decision tree, you start at the **root node** and follow the path based on the feature values of your example.

Let's trace the path:

1. **Root node** → First decision: **Ear shape**
   - The animal has **floppy ears**
   - Follow the "Floppy" branch

2. **Second level** → Next decision: **Face shape**
   - The animal has a **round face shape**
   - Follow the "Round" branch

3. **Third level** → Next decision: **Whiskers?**
   - The animal **has whiskers** (Present)
   - Follow the "Present" branch

4. **Leaf node** → This is a **cat** (the leaf node predicts "cat")

#### Visual Path

```
                  Ear shape
                 /    |    \
           Floppy  Pointy  ?
              |
          Face shape
           /      \
       Round    Not round
         |
     Whiskers?
      /     \
  Present  Absent
    |
  cat  <--- You are here
```

---

### Why "Not a cat" is Incorrect

- If you followed the path incorrectly (e.g., misread "floppy" as "pointy," or "round" as "not round," or "present" as "absent"), you would end at a different leaf node that might predict "not a cat."
- However, based strictly on the features given and following the tree correctly, the prediction is "cat."

---

### Real-World Example

**Concept: Decision Tree Prediction**

Think of this like a **flowchart for identifying animals**:

- **Step 1:** Ear shape? → Floppy (like a dog or cat)
- **Step 2:** Face shape? → Round (more cat-like than dog-like)
- **Step 3:** Whiskers? → Present (strong cat indicator)

Each question narrows down the possibilities until you reach a conclusion. This is exactly how decision trees work in practice, whether for medical diagnosis, customer segmentation, or spam detection.

## Summary Table: Key Concepts

| Concept | Definition | Real-World Example |
|---------|------------|-------------------|
| **Purity** | How homogeneous a node is (all examples belong to one class) | A perfectly sorted pile of apples vs. oranges |
| **Impurity** | How mixed the classes are in a node | A mixed pile of apples and oranges |
| **Splitting Criterion** | The rule used to choose which feature to split on | Choosing which medical test to run first |
| **Tree Traversal** | Following the path from root to leaf based on feature values | Using a flowchart to diagnose a problem |

---

## Key Takeaways

1. **Decision trees choose splits that maximize purity** in the resulting child nodes.
2. **Perfect separation** (100% pure children) is the ideal, but not always possible.
3. To make a prediction, **follow the path** from root to leaf based on the example's feature values.
4. Decision trees are intuitive and mimic human decision-making processes.
